```python
import newspaper
newspaper.languages()
```

```python
cbs_paper = newspaper.build('http://cbs.com', memoize_articles=False)
print (cbs_paper.size())
```

```python
# getting all categories 
cbs_paper.category_urls()
```

```python
#Extracting Source brand & description
print(cbs_paper.brand)
print(cbs_paper.description)
```

```python
# print all articles 
print(cbs_paper.articles[0].title)
print (len(cbs_paper.articles))
```

```python
#Downloading an Article
first_article = cbs_paper.articles[0]
first_article.download()
print(first_article.html)
```

```python
#Parsing an Article
first_article.parse()
print(first_article.text)
print(first_article.top_image)
print(first_article.authors)
print(first_article.title)
print(first_article.images[0])
print(first_article.movies)
```

```python
#Performing NLP on an Article
first_article.nlp()
print(first_article.summary)
print(first_article.keywords)
print(first_article.canonical_link)
```
